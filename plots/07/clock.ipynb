```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
plt.rcParams.update({'font.size': 12})
from scipy import stats
```

```python
ra = []
t1 = []
t2 = []
with open('clock.tsv', 'r') as f:
    while True:
        try:
            a,b,c = map(float, f.readline().strip().split('\t'))
            ra.append(a)
            t1.append(b)
            t2.append(c)
        except:
            break
#
slope, intercept, r_value, p_value, std_err = stats.linregress(ra, t1)
print('y = (%f) + (%f) * x' % (intercept, slope))
print('r^2 = %f' % r_value**2)

x = np.linspace(0, 410, 10)
y = intercept + slope * x
plt.plot(ra, t1, 'o', label=r'$t_1$')
plt.plot(ra, t2, 'o', label=r'$t_2$')
plt.plot(x, y, label='fitted')
plt.xlim(0, 410)
plt.ylim(0, 40)
plt.xlabel(r'$R_{A}$ (kÎ©)')
plt.ylabel(r'$t$ (ms)')
plt.title(r"Controling clock's period with $R_{A}$")
plt.hlines(6.46, 0, 500, label='6.46ms', linestyles='dashed')

plt.legend(loc=0)
plt.savefig('clock-plot.pdf')
```
