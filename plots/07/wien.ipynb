```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
plt.rcParams.update({'font.size': 12})
```

```python
rbra = []
ptp = []

with open('wien.tsv', 'r') as f:
    while True:
        try:
            a, b, c = map(float, f.readline().strip().split())
            rbra.append(a)
            ptp.append(c/2)
        except:
            break

#
plt.plot(rbra, ptp, 'o')
plt.xlim(2, 2.8)
plt.ylim(0, 15)
plt.title(r'$R_{b}/R_{a}$ vs. output signal amplitude')
plt.xlabel(r'$R_{b}/R_{a}$')
plt.ylabel('Amplitude of output signal (V)')

plt.savefig('rbra.pdf')
```
