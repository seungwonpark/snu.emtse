```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
```

```python
d = []
v = [] # minimal voltage
with open('dist.tsv', 'r') as f:
    while True:
        try:
            a, b = map(float, f.readline().strip().split('\t'))
            d.append(a / 100)
            v.append(b / 750 * 1000)
        except:
            break

plt.title('Photo diode\'s reaction to luminance')
plt.xlabel('Distance (m)')
plt.ylabel('Minimal current (mA)')
plt.plot(d, v, 'o')
plt.savefig('photodiode-result.pdf')
```
