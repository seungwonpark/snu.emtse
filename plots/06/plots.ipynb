```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 12})
```

```python
freq_diff = []
g_diff = []
g_diff_deno = []

with open('diff.tsv', 'r') as f:    
    while True:
        try:
            a,_,b,_,c = map(float, f.readline().strip().split('\t'))
            freq_diff.append(a)
            g_diff.append(b)
            g_diff_deno.append(c)
        except:
            break

freq_inte = []
g_inte = []
g_inte_deno = []
vin_inte = 1.92

with open('inte.tsv', 'r') as f:
    while True:
        try:
            a, b, c, d = map(float, f.readline().strip().split('\t'))
            freq_inte.append(a)
            g_inte.append(b / vin_inte)
            g_inte_deno.append(d / vin_inte)
        except:
            break
```

```python
plt.plot(freq_inte, g_inte, 'o')
plt.title('Integrator frequency vs. Gain')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('f (Hz)')
plt.ylabel('gain')
plt.savefig('integrator.pdf')
```

```python
plt.plot(freq_inte, g_inte_deno, 'o')
plt.title(r'Integrator + $R_s$ frequency vs. Gain')
plt.xscale('log')
# plt.yscale('log')
plt.xlabel('f (Hz)')
plt.ylabel('gain')
plt.savefig('integrator+rs.pdf')
```

```python
plt.plot(freq_diff, g_diff, 'o')
plt.title('Differentiator frequency vs. Gain')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('f (Hz)')
plt.ylabel('gain')
plt.savefig('differentiator.pdf')
```

```python
plt.plot(freq_diff, g_diff_deno, 'o')
plt.title(r'Differentiator + $ R_s $ frequency vs. Gain')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('f (Hz)')
plt.ylabel('gain')
plt.savefig('differentiator+rs.pdf')
```

```python
plt.plot(freq_diff, g_diff, 'o', label='differentiator')
plt.plot(freq_inte, g_inte, 'o', label='integrator')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('f (Hz)')
plt.ylabel('gain')
plt.title('gain(f) without denoising')
plt.hlines(14.0, 10**1, 10**5, label=r'$V_{\rm out}$=27V', linestyles='dashed')
plt.xlim(10**1, 2*10**4)
plt.legend(loc=6)
plt.savefig('diff+inte_without_denoising.pdf')
```

```python
plt.plot(freq_diff, g_diff_deno, 'o', label='differentiator')
plt.plot(freq_inte, g_inte_deno, 'o', label='integrator')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('f (Hz)')
plt.ylabel('gain')
plt.title('gain(f) with denoising')
plt.legend(loc=0)
plt.savefig('diff+inte_with_denoising.pdf')
```
